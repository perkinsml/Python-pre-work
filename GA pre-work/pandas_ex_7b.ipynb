```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame([["Europe", "UK", 100000], ["Europe", "DE", 20000], ["Africa", "Kenya", 40000], ["Africa", "Morocco", 20000], ["Africa", "Chad", 10000]], columns=["Continent","Country","Pageviews"])
```

```python
df.plot(x='Country', y='Pageviews', kind='bar')
```

```python
df.plot.bar(x="Country", y="Pageviews")
```

```python
df.plot.bar(x="Country", y="Pageviews", figsize=(8,4), title = "Pageviews by country")
```

```python
df.plot.pie(x='Country', y="Pageviews", legend="True")
```
